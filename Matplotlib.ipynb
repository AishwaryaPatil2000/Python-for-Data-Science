Matplotlib Inline is a package that supports Matplotlib to display plots directly inline and save them to notebooks.


This line tells Jupyter to display Matplotlib plots directly in the output of the notebook. One of the major benefits of using "Matplotlib inline" is that it allows you to quickly and easily visualize your data without having to save plots as image files or open them in a separate window.

#Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#defining the dataset
x=np.arange(0,10,0.1)
y=2*x +5

#plotting
plt.plot(x,y)
plt.show()
```

```python
#Customizing line plot
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#defining the dataset
x=np.arange(0,10,1)
y=2*x +5

#plotting
plt.plot(x,y,linewidth =2.0,linestyle=':',alpha =0.5,color = 'r',marker ='o')
plt.title("Line Plot")
plt.xlabel("x-Axis")
plt.ylabel("y-Axis")
plt.legend(['line1'],loc='best')     #legend is used for line differentiation when we have multiple lines
plt.grid(True)
plt.show()
```

```python
#changing the gifure size


#Customizing line plot
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#defining the dataset
x=np.arange(0,10,1)
y=2*x +5

#changing fig size
fig=plt.figure(figsize=(10,5))

#plotting
plt.plot(x,y,linewidth =2.0,linestyle=':',alpha =0.5,color = 'r',marker ='o')
plt.title("Line Plot")
plt.xlabel("x-Axis")
plt.ylabel("y-Axis")
plt.legend(['line1'],loc='best')     #legend is used for line differentiation when we have multiple lines
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x=np.arange(0,10,1)
y1=x*2 + 5
y2=x*3 + 10

#creating 2 or more plots in a fig
plt.subplot(1,2,1)     #(rows,cols,index)
plt.plot(x,y1)
plt.title('graph1')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('graph2')

plt.show()
```

#Bar Plot

```python
data={'apples':20 ,'Mangoes':15,'lemon':30,'Oranges':10}
names=list(data.keys())
values=list(data.values())

fig=plt.figure(figsize=(10,5))
plt.bar(names,values)
plt.show()
```

```python
data={'apples':20 ,'Mangoes':15,'lemon':30,'Oranges':10}
names=list(data.keys())
values=list(data.values())

fig=plt.figure(figsize=(10,5))
plt.bar(names,values,color="orange")
plt.title("Bar Graph")
plt.xlabel('fruits')
plt.ylabel('quantity')
plt.show()
```

```python
plt.barh(names,values,color='blue')
plt.title("Bar Graph")
plt.xlabel('fruits')
plt.ylabel('quantity')
plt.show()
```

#Scatter Plot

```python
a=[10,20,30,40,50,60,70,80]
b=[5,4,3,2,6,1,4,2]
x=[1,2,3,4,3,2,1,2]
plt.scatter(a,b)
plt.scatter(a,x)
plt.show
```

```python
#Customising scatter plot
a=[10,20,30,40,50,60,70,80]
b=[5,4,3,2,6,1,4,2]
x=[1,2,3,4,3,2,1,2]
plt.scatter(a,b,c='g',s=300,edgecolors='y',marker='o',alpha=0.5)
plt.scatter(a,x,c='r',s=400,edgecolors='b',marker='4',alpha=1)
plt.legend(['b','x'],loc='best')
plt.title("Scatter plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.savefig('scatterplt.png')
plt.savefig('scatterplt.jpeg')
plt.grid()
plt.show()
```

#Histogram

#Box Plot

```python
Total=[20,4,1,30,20,10,20,70,30,10]
order=[10,3,1,15,17,2,30,44,2,1]
discount=[30,10,20,5,10,20,50,60,20,45]
data=list([Total,order,discount])

plt.boxplot(data,showmeans=True)
plt.title("box plot")
plt.show()
```

#Violin Plot

```python
#Generally used for huge dataset
Total=[20,4,1,30,20,10,20,70,30,10]
order=[10,3,1,15,17,2,30,44,2,1]
discount=[30,10,20,5,10,20,50,60,20,45]
data=list([Total,order,discount])

plt.violinplot(data,showmeans=True,showmedians=True)
plt.title("Violin plot")
plt.show()
```

#Pie Chart

```python
label=['Dog','Cat','Wolf','Lion']
Sizes=[50,45,60,80]

plt.pie(Sizes,labels=label,autopct='%1.2f%%',startangle=90,shadow=True)
plt.title("pie chart")
plt.show()
```

#Doughnut Chart

```python
group_names=['groupA','GroupB','groupC']
group_size=[20,30,50]
size_centre=[5]

pie1=plt.pie(group_size,labels=group_names,radius=1.5)
pie2=plt.pie(size_centre,radius=1.0,colors='w')
plt.show()
```

```python
#customising doughnut chart
group_names=['groupA','GroupB','groupC']
group_size=[20,30,50]
size_centre=[5]
a,b,c =[plt.cm.Blues,plt.cm.Reds,plt.cm.Greens]
pie1=plt.pie(group_size,labels=group_names,radius=1.5,colors=[a(0.5),b(1.5),c(1.0)])
pie2=plt.pie(size_centre,radius=1.0,colors='w')
plt.show()
```

#Area plot

```python
x=range(1,15)
y=[1,4,6,8,4,5,3,2,4,1,5,6,8,7]

plt.stackplot(x,y)
#merge area chart with line plot
plt.plot(x,y,color='Black')
plt.show()
```
