```python
#Memory size

import numpy as np
import sys

#define a list
l=range(1000)
print(sys.getsizeof(1)*len(l))

#define a numpy array
a=np.array(1000)
print(a.size*a.itemsize)
```

```python
#Speed
import time
def using_list():
  t1=time.time()
  x=range(1000)
  y=range(1000)
  z=[x[i]+y[i] for i in range(len(x))]
  return time.time()-t1

def using_numpy():
  t2=time.time()
  x=np.arange(1000)
  y=np.arange(1000)
  z=x+y                                           #more convinient
  return time.time()-t2

list_time=using_list()
numpy_time=using_numpy()

print(list_time,numpy_time)
print("numpy array is " +str(numpy_time/list_time) + "times faster" )
```
