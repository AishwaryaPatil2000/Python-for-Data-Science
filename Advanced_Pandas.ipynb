#CSV

```python
from io import StringIO ,BytesIO
```

```python
data = ('col1,col2,col3\n'
            'x,y,1\n'
            'a,b,2\n'
            'c,d,3')
```

```python
type(data)
```

```python
df1=pd.read_csv(StringIO(data))
```

```python
df1.head()
```

```python
df1=pd.read_csv(StringIO(data),usecols=['col1','col3'])
df1
```

```python
data2 = ('a,b,c,d\n'
            '1,2,3,4\n'
            '5,6,7,8\n'
            '9,10,11')
```

```python
print(data2)
```

```python
df3=pd.read_csv(StringIO(data2),dtype=object)
```

```python
df3
```

```python
df4=pd.read_csv(StringIO(data),dtype={'b':int,'c':float,'a':'Int64'})
df4
```

```python
df4.dtypes
```

```python
data3 = ('index,a,b,c\n'
           '4,apple,bat,5.7\n'
            '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data3))
```

```python
df5=pd.read_csv(StringIO(data3),usecols=['b','c'],index_col=False)
df5
```

```python
## Quoting and Escape Characters¶. Very useful in NLP

data4 = 'a,b\n"hello, \\"Bob\\", nice to see you",5'
```

```python
df6=pd.read_csv(StringIO(data4),escapechar='\\')
df6
```
