#Merge Operation

```python
player=['p1','p2','p3']
point=[8,9,6]
title=['game1','game2','game3']
df=pd.DataFrame({'player':player,'points':point,'title':title})
df
```

```python
player=['p1','p5','p6']
power=['punch','kick','elbow']
title=['game1','game5','game6']
df1=pd.DataFrame({'player':player,'power':power,'title':title})
df1
```

```python
#Inner Merge ---returns commonvalues  based on common columns present in two df    Default merge
df.merge(df1, on='player' ,how='inner')
```

```python
df.merge(df1, on='player' ,how='left')
```

```python
df.merge(df1, on='player' ,how='right')
```

```python
df.merge(df1, on='player' ,how='outer')
```

#Join operation

```python
player=['p1','p2','p3']
point=[8,9,6]
title=['game1','game2','game3']
df2=pd.DataFrame({'player':player,'points':point,'title':title},index=['L1','L2','L3'])
df2
```

```python
player=['p1','p5','p6']
power=['punch','kick','elbow']
title=['game1','game5','game6']
df3=pd.DataFrame({'players':player,'power':power,'titles':title},index=['L2','L3','L4'])
df3
```

```python
df2.join(df3 ,how='inner')   #join is performed based on indexing
```

```python
df2.join(df3 ,how='left')
```

```python
df2.join(df3 ,how='right')
```

```python
df2.join(df3 ,how='outer')
```

```python
#concatenate df
pd.concat([df3,df2])
```
